# Plotting example

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.0, 4, 300)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.sin(3 * np.pi * x))
ax.set_xlim([0.0, 4])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.sin(3 * np.pi * x) 
            + (1 / (x + 0.1)) * np.sin(20 * np.pi * x))
ax.set_xlim([0.0, 4])
plt.show()
```
